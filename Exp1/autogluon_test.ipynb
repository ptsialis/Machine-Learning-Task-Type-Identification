```python
import pandas as pd
from autogluon.global_controller import infer_problem_type
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score,confusion_matrix,balanced_accuracy_score
from pathlib import Path
import plotly.figure_factory as ff
pd.set_option('display.max_rows', 500)
```

```python
df_autogluon_test = pd.read_csv("../meta_datasets/tabular_meta_datasets/exp1_test_autogluon.csv")
```

```python
def map_problem_type(problem_type: str, sub_problem_type: str) -> str:
    """
    Maps problem_type and sub_problem_type to a standardized string.
    """

    # normalize inputs
    problem_type = problem_type.strip().lower()
    sub_problem_type = sub_problem_type.strip().lower()

    # explicit tabular handling
    if problem_type == "tabular":
        if sub_problem_type in {"binary", "binary classification"}:
            return "TABULAR_CLASSIFICATION_BINARY"
        elif sub_problem_type in {"multiclass", "multiclass classification"}:
            return "TABULAR_CLASSIFICATION_MULTICLASS"
        elif sub_problem_type == "regression":
            return "TABULAR_REGRESSION"

    raise ValueError(
        f"Unsupported combination: "
        f"problem_type='{problem_type}', sub_problem_type='{sub_problem_type}'"
    )
```

```python
def map_autogluon_problem_types(problem_types):
    """
    Maps AutoGluon problem types to tabular task categories.

    Parameters
    ----------
    problem_types : list[str]
        List of problem types inferred by AutoGluon
        (e.g. 'binary', 'multiclass', 'regression').

    Returns
    -------
    list[str]
        Mapped tabular task types.
    """
    mapping = {
        "binary": "TABULAR_CLASSIFICATION_BINARY",
        "multiclass": "TABULAR_CLASSIFICATION_MULTICLASS",
        "regression": "TABULAR_REGRESSION",
    }

    return [mapping[p] for p in problem_types]
```

```python
df_autogluon_test["ground_truth"] = df_autogluon_test.apply(
    lambda row: map_problem_type(row["problem_type"], row["sub_problem_type"]),
    axis=1
)
```

# Loop through dataset to get autogluon prediction results

```python
prediction_autogluon = []
```

```python
for index,row in df_autogluon_test.iterrows():
    current_dataset = pd.read_csv(row["file_path"])
    problem_type= infer_problem_type(current_dataset[row["target_variable"]])
    prediction_autogluon.append(problem_type)
```

```python
prediction_autogluon
```

```python
results_df = pd.DataFrame()
results_df["gt"] = df_autogluon_test["sub_problem_type"].to_list()
results_df["pred"] = prediction_autogluon
y_true= df_autogluon_test["sub_problem_type"].to_list()
y_hat = prediction_autogluon
b_acc = balanced_accuracy_score(y_true,y_hat)
acc=accuracy_score(y_true,y_hat)
precision=precision_score(y_true,y_hat,average="macro")
recall=recall_score(y_true,y_hat,average="macro")
f1=f1_score(y_true,y_hat,average="macro")
```
